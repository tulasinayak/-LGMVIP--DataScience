```python
import cv2 
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("golden-retriever-dog.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
#converting to grey scale image
img_grey=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img_grey)
```

```python
#inverting the image
img_inv = cv2.bitwise_not(img_grey)
plt.imshow(img_inv)
```

```python
#blurring the image using gaussian blur
img_smooth = cv2.GaussianBlur(img_inv,(21,21),0)
plt.imshow(img_smooth)
```

```python
img_blur_inv=cv2.bitwise_not(img_smooth)
plt.imshow(img_blur_inv)
```

```python
final_img=cv2.divide(img_grey,img_blur_inv,scale=256.0)
```

```python
final=cv2.resize(final_img,(540,540))
cv2.imshow('final image',final)
cv2.waitKey()
```
